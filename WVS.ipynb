```python
import json
import csv

with open('WVS/data.csv', mode='r', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    data = [row for row in reader]
```

```python
q_list = [f"Q{i}" for i in range(176, 199)] + [f"Q{i}" for i in range(1, 45 + 1)]
```

## Pre-process

```python
new_data = []
for d in data:
    nd = {k: v for k, v in d.items() if k.startswith('Q')}
    new_data.append(nd)
```

```python
men = []
women = []
for d in new_data:
    nd = {}
    for key in d.keys():
        if key in q_list:
            nd[key] = d[key]
    if d['Q260'] == '1':
        men.append(nd)
    elif d['Q260'] == '2':
        women.append(nd)
```

```python
from collections import defaultdict

sum_dict = defaultdict(float)
count_dict = defaultdict(int)

for sample in women:
    for key, value in sample.items():
        if list(new_data[0].keys()).index(key) >= 296:
            break
        sum_dict[key] += int(value)
        count_dict[key] += 1

women_avg = {key: sum_dict[key] / count_dict[key] for key in sum_dict}
```

```python
sum_dict = defaultdict(float)
count_dict = defaultdict(int)

for sample in men:
    for key, value in sample.items():
        if list(new_data[0].keys()).index(key) >= 296:
            break
        sum_dict[key] += int(value)
        count_dict[key] += 1

men_avg = {key: sum_dict[key] / count_dict[key] for key in sum_dict}
```

```python
relative_threshold = 0.04

very_different_keys = []

for key in men_avg:
    if key in women_avg:
        a = men_avg[key]
        b = women_avg[key]
        if a == 0 and b == 0:
            continue  # no difference
        elif a == 0 or b == 0:
            very_different_keys.append(key)  # extreme difference
        else:
            relative_diff = abs(a - b) / max(abs(a), abs(b))
            if relative_diff > relative_threshold:
                very_different_keys.append(key)

```

```python
from collections import defaultdict

def convert_samples_to_question_lists(samples):
    question_responses = defaultdict(list)
    
    for sample in samples:
        for question, answer in sample.items():
            question_responses[question].append(int(answer))  # convert to int if needed

    return dict(question_responses)
```

```python
def format_list(input_str):
    lines = input_str.strip().split('\n')
    result = [None] * len(lines)

    for line in lines:
        number_str, value_str = line.split('. ')
        index = int(number_str) - 1
        value = int(value_str)  # convert to int; remove if you want strings
        result[index] = value
    return result
```

```python
import json

with open('completed_results/wvs_men.json','r') as file:
    men = json.load(file)
with open('completed_results/wvs_women.json','r') as file:
    women = json.load(file)
```

```python
with open('completed_results/wvs_qwen.json','r') as file:
    llm_responses = json.load(file)
```

```python
from statistics import mean

alignment = {}
threshold_ratio = 0.05

for q in llm_responses:
    llm_val = llm_responses[q]
    male_avg = mean(men[q])
    female_avg = mean(women[q])
    
    dist_to_male = abs(llm_val - male_avg)
    dist_to_female = abs(llm_val - female_avg)
    
    min_dist = min(dist_to_male, dist_to_female)
    diff = abs(dist_to_male - dist_to_female)

    if diff <= threshold_ratio * min_dist:
        alignment[q] = 'neither'
    elif dist_to_male < dist_to_female:
        alignment[q] = 'men'
    else:
        alignment[q] = 'women'
```

```python
with open("completed_results/wvs_align_qwen.json", "w") as json_file:
  json.dump(alignment, json_file, indent=4)
```

```python
import requests
import time
from tqdm import tqdm

url = ""
MODEL = "Qwen/Qwen2.5-7B-Instruct"
```

```python
results = []

counter = 1
for sample in tqdm(data):
    user_input = sample

    payload = {
        "model": MODEL,
        "stream": False,
        "max_tokens": 250,
        "min_p": 0.05,
        "temperature": 0.7,
        "top_p": 0.7,
        "top_k": 50,
        "frequency_penalty": 0.5,
        "n": 1,
        "messages": user_input
    }
    headers = {
        "Authorization": "",
        "Content-Type": "application/json"
    }

    response = requests.request("POST", url, json=payload, headers=headers).json()
    pred = response['choices'][0]['message']['content']
    results.append({
                "qid": counter,
                "response": pred
            })
    counter += 1
```

```python
with open("wvs_"+MODEL+".json", "w") as json_file:
  json.dump(results, json_file, indent=4)
```

```python
import matplotlib.pyplot as plt

def plot_lines(data, color=None):
    categories = ["men", "neither", "women"]
    color = color
    x_pos = {cat: i for i, cat in enumerate(categories)}

    # Prepare figure
    fig, ax = plt.subplots(figsize=(3, 4))

    # Plot 3 vertical lines
    for x in x_pos.values():
        ax.vlines(x, ymin=0, ymax=len(data)-1, color='lightgray', linewidth=2)

    # Plot red and blue dots
    if color:
        for i, category in enumerate(data.values()):
            x = x_pos[category]
            ax.plot(x, i, 'o', color=color)
    else:
        for i, category in enumerate(data.values()):
            x = x_pos[category]
            color = 'ro' if i < 45 else 'bo'
            ax.plot(x, i, color)  # red dot

    # Format axes
    ax.set_xticks(list(x_pos.values()))
    ax.set_xticklabels(categories)
    ax.set_yticks([])  # remove y-axis ticks
    ax.set_yticklabels([])  # remove y-axis labels
    ax.set_xlim(-0.5, len(categories) - 0.5)
    ax.invert_yaxis()

    # Remove frame (spines)
    for spine in ax.spines.values():
        spine.set_visible(False)

    plt.tight_layout()
    plt.show()
```
